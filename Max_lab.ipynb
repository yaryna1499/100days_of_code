```python
import numpy as np
import math
import tkinter as tk
root = tk.Tk ()
root
```

```python
n = 7 # число лінійок
m = 7 # число стовпців
p = 0.7
sum = 0 #початкове значення суми елементів матриці
r = np.random.randint(0,2,(n,m)) # матриця координат
s = np.random.randint(0,2,(n,m)) # матриця зайнятих комірок
x0 = 10
y0 = 10
dx = 20
dy = 20
x1 = 30
y1 = 30
root = tk.Tk ()
can = tk.Canvas(root,width=500, height=500)
can.pack() #
for i in range(n):
    for j in range(m):
            if r[i][j] < p:
                s[i][j] = 1
                square = can.create_rectangle(x0, y0, x1, y1, fill="green")
                text = can.create_text((x0+x1)/2,(y0+y1)/2, text="1",font=("Helvectica", "16"))
            else:
                s[i][j] = 0
                square = can.create_rectangle(x0, y0, x1, y1, fill='white')
                text = can.create_text((x0+x1)/2,(y0+y1)/2, text="0",font=("Helvectica", "16"))
            y0 += dy
            y1 += dy
    y0 = 10
    y1 = 30
    x0 += dx
    x1 += dx
root.mainloop()
for i in range(0, n):
    for j in range(0, m):
        sum = sum + s[i][j]
        sum1 = float(sum)
        number = float(n*m)
        percol = sum1/number
print ('summa = %s persolation=%.2f' %(sum1,percol))
```
